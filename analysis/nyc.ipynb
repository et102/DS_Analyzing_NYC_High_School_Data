### Read in data from local repo

```python
import pandas as pd
import numpy
import re
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# read in all .csv file

data_files = [
    "ap_2010.csv",
    "class_size.csv",
    "demographics.csv",
    "graduation.csv",
    "hs_directory.csv",
    "sat_results.csv"
]

data = {}

for f in data_files:
    d = pd.read_csv("../data/{0}".format(f))
    data[f.replace(".csv", "")] = d
```

### Read in survey file (*.txt)

```python
# read in survey file
all_survey = pd.read_csv("../data/survey_all.txt", delimiter="\t", encoding='windows-1252')
d75_survey = pd.read_csv("../data/survey_d75.txt", delimiter="\t", encoding='windows-1252')

#Concatenating
survey = pd.concat([all_survey, d75_survey], axis=0)

# convert lowercase to upcase
survey["DBN"] = survey["dbn"]

# filter columns
survey_fields = [
    "DBN", 
    "rr_s", 
    "rr_t", 
    "rr_p", 
    "N_s", 
    "N_t", 
    "N_p", 
    "saf_p_11", 
    "com_p_11", 
    "eng_p_11", 
    "aca_p_11", 
    "saf_t_11", 
    "com_t_11", 
    "eng_t_11", 
    "aca_t_11", 
    "saf_s_11", 
    "com_s_11", 
    "eng_s_11", 
    "aca_s_11", 
    "saf_tot_11", 
    "com_tot_11", 
    "eng_tot_11", 
    "aca_tot_11",
]
survey = survey.loc[:,survey_fields]
data["survey"] = survey
```

### Add DBN columns to class_size data

```python
# revise dbn column in "hs_directory"
data["hs_directory"]["DBN"] = data["hs_directory"]["dbn"]

## Pad CSD column to creat DBN column: DBN = CSD + 
## zfill(): specify width and append 0
def padded(temp):

    temp_string = str(temp)
    
    if len(temp_string) > 1:
       return temp_string
    else: 
       return temp_string.zfill(2)
      
data["class_size"]["padded_csd"]=data["class_size"]["CSD"].apply(padded)
    
data["class_size"]["DBN"] = data["class_size"]["padded_csd"] + data["class_size"]["SCHOOL CODE"]

#print(data["class_size"].head())
```

### Convert columns to numeric and Map extraction

```python
cols = ['SAT Math Avg. Score', 'SAT Critical Reading Avg. Score', 'SAT Writing Avg. Score']

# pd.to_numeric(errors="coerce"): pandas treats any invalid strings it can't convert to numbers as missing values instead

for c in cols:
    data["sat_results"][c] = pd.to_numeric(data["sat_results"][c], errors="coerce")

data['sat_results']['sat_score'] = data['sat_results'][cols[0]] + data['sat_results'][cols[1]] + data['sat_results'][cols[2]]

## Position extration for map
## re.findall( pattern, sring)

def find_lat(loc):
    coords = re.findall("\(.+, .+\)", loc)
    lat = coords[0].split(",")[0].replace("(", "")
    return lat

def find_lon(loc):
    coords = re.findall("\(.+, .+\)", loc)
    lon = coords[0].split(",")[1].replace(")", "").strip()
    return lon

data["hs_directory"]["lat"] = data["hs_directory"]["Location 1"].apply(find_lat)
data["hs_directory"]["lon"] = data["hs_directory"]["Location 1"].apply(find_lon)

data["hs_directory"]["lat"] = pd.to_numeric(data["hs_directory"]["lat"], errors="coerce")
data["hs_directory"]["lon"] = pd.to_numeric(data["hs_directory"]["lon"], errors="coerce")
```

### Condense dataset

```python
## filter class_size with "Grade" and "Program Type"
class_size = data["class_size"]
class_size = class_size[class_size["GRADE "] == "09-12"]
class_size = class_size[class_size["PROGRAM TYPE"] == "GEN ED"]

## groupby DBN and calculate mean_size
## reset index
class_size = class_size.groupby("DBN").agg(numpy.mean)
class_size.reset_index(inplace=True)
data["class_size"] = class_size

## filter "schoolyear"
data["demographics"] = data["demographics"][data["demographics"]["schoolyear"] == 20112012]

## filter "graduation"
data["graduation"] = data["graduation"][data["graduation"]["Cohort"] == "2006"]
data["graduation"] = data["graduation"][data["graduation"]["Demographic"] == "Total Cohort"]
```

### Convert AP scores to numeric

```python
cols = ['AP Test Takers ', 'Total Exams Taken', 'Number of Exams with scores 3 4 or 5']

for col in cols:
    data["ap_2010"][col] = pd.to_numeric(data["ap_2010"][col], errors="coerce")
```

### Combine the datasets

```python
combined = data["sat_results"]

# pd.merge = Join in SQL
combined = combined.merge(data["ap_2010"], on="DBN", how="left")
combined = combined.merge(data["graduation"], on="DBN", how="left")

to_merge = ["class_size", "demographics", "survey", "hs_directory"]

for m in to_merge:
    combined = combined.merge(data[m], on="DBN", how="inner")

# currently looking for high-level correlations,so replacing null values with mean.   
combined = combined.fillna(combined.mean())
combined = combined.fillna(0)
combined
```

### Add a school district column for mapping

```python
def get_first_two_chars(dbn):
    return dbn[0:2]

combined["school_dist"] = combined["DBN"].apply(get_first_two_chars)
```

### Find correlations

```python
correlations = combined.corr()
correlations = correlations["sat_score"]
print(correlations)
```

Plot correlation with Bar chart: pandas.DataFrame.plot.bar()

```python
#Remove DBN since it's a unique identifier, not a useful numerical value for correlation.
survey_fields.remove("DBN")

%matplotlib inline
combined.corr()["sat_score"][survey_fields].plot.bar()
```

### Findings:  

saf_t_11 and saf_s_11, which refer to safety peceived by teachers, are highly correlated with SAT score

### Plotting safety

```python
## Aggregation combined with "school_dist" and calculate mean
## Assign back to dist_agg

dist_agg = combined.groupby("school_dist").agg(numpy.mean)
dist_agg.reset_index(inplace=True)
dist_agg
```

```python

m = Basemap(
    projection='merc', 
    llcrnrlat=40.496044, 
    urcrnrlat=40.915256, 
    llcrnrlon=-74.255735, 
    urcrnrlon=-73.700272,
    resolution='i'
)

m.drawmapboundary(fill_color='#85A6D9')
m.drawcoastlines(color='#6D5F47', linewidth=.4)
m.drawrivers(color='#6D5F47', linewidth=.4)

longitudes = dist_agg["lon"].tolist()
latitudes = dist_agg["lat"].tolist()
m.scatter(longitudes, latitudes, s=50, zorder=2, latlon=True, c=dist_agg["saf_s_11"], cmap="summer")
plt.show()
```

### Findings:  

Upper Manhattan and parts of Queens and the Bronx tend to have higher safety scores, whereas Brooklyn has low safety scores.
